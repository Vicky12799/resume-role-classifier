```python
import re
def clean_function(resumeText):
    resumeText = re.sub('http\S+\s*', ' ', resumeText)  # remove URLs
    resumeText = re.sub('RT|cc', ' ', resumeText)  # remove RT and cc
    resumeText = re.sub('#\S+', '', resumeText)  # remove hashtags
    resumeText = re.sub('@\S+', '  ', resumeText)  # remove mentions
    resumeText = re.sub('[%s]' % re.escape("""!"#$%&'()*+,-./:;<=>?@[\]^_`{|}~"""), ' ', resumeText)  # remove punctuations
    resumeText = re.sub(r'[^\x00-\x7f]',r' ', resumeText) 
    resumeText = re.sub('\s+', ' ', resumeText)  # remove extra whitespace
    return resumeText
```

```python
df1= df.copy()
df1
```

```python
df1['cleaned_resume'] = df1['Resume'].apply(lambda x: clean_function(x))
df1.head()
```

```python
import spacy
from spacy.lang.en.stop_words import STOP_WORDS
```

```python
nlp = spacy.load("en_core_web_sm")
```

```python
len(STOP_WORDS)
```

```python
def preprocess(text):
    doc = nlp(text)
    no_stop_words = [token.text for token in doc if not token.is_stop and not token.is_punct]
    return " ".join(no_stop_words)
```

```python
df1["preprocessed_resume"] = df1["cleaned_resume"].apply(preprocess)
```

```python
df2 = df1.copy()
df2
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
label_encoder = LabelEncoder()
```

```python
df2["encoded_category"] = label_encoder.fit_transform(df2["Category"])
```

```python
df2
```

```python
import pickle
with open('./model/label_encoder.pkl', 'wb') as le_file:
    pickle.dump(label_encoder, le_file)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(df2.preprocessed_resume,df2.encoded_category,test_size = 0.2)
```

```python
type(X_train.values)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
v = CountVectorizer()
```

```python
X_train_cv = v.fit_transform(X_train.values)
X_train_cv
```

```python
X_train_cv.toarray()
```

```python
v.get_feature_names_out().shape
```

```python
v.vocabulary_
```

```python
from sklearn.pipeline import Pipeline
from sklearn.naive_bayes import MultinomialNB
```

```python
clf = Pipeline([
    ('vectorizer', CountVectorizer()),
    ('nb',MultinomialNB())
])
```

```python
clf.fit(X_train,y_train)
```

```python
clf.score(X_test,y_test)
```

```python
y_pred = clf.predict(X_test)
```

```python
from sklearn.metrics import classification_report
classification_report(y_test,y_pred)
```
